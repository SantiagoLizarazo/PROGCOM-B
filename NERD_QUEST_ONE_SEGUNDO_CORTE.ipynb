##Reactivo Limitante.
En una reacción quimica, los reactivos pueden estar o no en la proporción exacta que determinan sus coeficientes estequiometricos.

###Ejemplo 1:
Tenemos 10 moles de H2 y 7 moles de O2 para formar agua.

###Ejemplo 2:
2H2(g) + O2(g) → 2H2O(l)

En resumen, un reactivo limitante es aquel que se consume primero y limita la cantidad de producto que se forma.

##Pureza de Reactivos y Productos.
La pureza de una sustancia radica en su porcentaje peso peso(%m/m).

###Eficiencia o Rendimiento de una Reacción

*   Rendimiento Teorico: Cantidad de producto, que según los calculos, se forman cuan reacciona todo el reactivo limitante.
*   Rendimiento Real: Cantidad de producto que realmente se forma en la reacción.

###Porcentaje de Rendimiento
El porcentaje de rendimiento es la relación entre el rendimiento real y el rendimiento teorico multiplicado por 100.

P.B = Rendimiento Real / Rendimiento Teorico x 100

##Equilibrio Quimico
El equilibrio es un estado en el que no hay muchos cambios notables.

*   Las velocidades de las reacciones directas e inversas se igualan.
*   Las concentraciones de los reactivos y productos permanecen constantes.

###Ejemplo.

*   Equilibrio Fisico: H20(l)⇌H20(g)
*   Elemento de lista: N204(g)⇌2NO2(g)
