```python
#Funciones de Definición: Techo y Estructura Principal
def describir_techo():
    return {
        "material": "tejas de barro artesanales",
        "color": "amarillo",
        "estado": "desgaste por el sol y la lluvia",
        "caracteristica": "grande"
    }

def describir_paredes():
    return {
        "cantidad": 4,
        "material": "ladrillo rojo y cemento fino",
        "proteccion": "relieve en espiral (más conocido como consertina) en la parte superior para evitar robos"
    }

def describir_fachada():
    return {
        "color": "amarillo",
        "puerta": {
            "color": "rosada",
            "ancho_cm": 72.5,
            "alto_cm": 203,
            "material": "madera de roble con herrajes de hierro forjado"
        }
    }

# Funciones de Definición: Lado Izquierdo
def describir_lateral_izquierdo():
    return {
        "color": "azul",
        "ventana": {
            "tipo": "translúcida",
            "material": "vidrio templado con marco de madera"
        },
        "logo_floristeria": {
            "nombre": "Flor by Ista",
            "diseño": "moderno y minimalista en letras cursivas",
            "elementos_visuales": "un racimo de 3 flores y dos hojas"
        }
    }

# Funciones de Definición: Parte Posterior
def describir_parte_posterior():
    return {
        "color": "amarillo",
        "tragaluz": "un traga luz para darle escencia de vida y naturalidad a la tienda",
        "matera_colgante": {
            "medidas_cm": "100 cm de ancho x 50 cm de alto",
            "material": "adobe y madera de roble",
            "contenido": "mezcla de abono y tierra húmeda",
            "flores_cantidad": 9,
            "flores_tipo": "girasoles"
        }
    }

# Funciones de Definición: Lado Derecho
def describir_lateral_derecho():
    return {
        "color": "azul",
        "ventana": {
            "tipo": "translúcida",
            "material": "vidrio templado y madera",
            "decoracion_interior": "plantas colgantes"
        },
        "estantes": {
            "tipos": ["espiral", "alargado", "sencillo"],
            "total_flores_venta": 3 + 3 + 8, # Lirio (3) + Orquídea (3) + Rosa (8)
            "detalle": {
                "espiral": {"items": 3, "tipo_flor": "lirio", "material_matero": "plástico"},
                "alargado": {"items": 3, "tipo_flor": "orquídea", "material_estante": "madera de roble"},
                "sencillo": {"items": 4, "tipo_flor": "rosa", "cantidad_rosas": 8, "material_matero": "plástico"}
            }
        }
    }

# Función Principal
def generar_descripcion_completa():

    techo = describir_techo()
    paredes = describir_paredes()
    fachada = describir_fachada()
    lateral_izq = describir_lateral_izquierdo()
    posterior = describir_parte_posterior()
    lateral_der = describir_lateral_derecho()

    total_flores_venta = lateral_der['estantes']['total_flores_venta']
    total_flores_jardin = posterior['matera_colgante']['flores_cantidad']

    # La descripción
    descripcion = "--- Descripción Arquitectónica de la Floristería Flor by Ista ---\n\n"

    # Techo y Estructura
    descripcion += f"Estructura Base: 4 paredes Hechas de {paredes['material']}, con un diseño de seguridad de {paredes['proteccion']}.\n"
    descripcion += f"Techo: De tamaño {techo['caracteristica']} realizado con {techo['material']}. Cuentan con un color {techo['color']} debido al {techo['estado']}.\n"

    # Fachada Principal
    descripcion += f"\nLa fachada principal, cuenta con un color {fachada['color']}:\n"
    descripcion += f" - La puerta de entrada es de color {fachada['puerta']['color']}, contando con unas medidas de ({fachada['puerta']['ancho_cm']} x {fachada['puerta']['alto_cm']} cm), fabricada en {fachada['puerta']['material']}.\n"

    # Lateral Izquierdo
    descripcion += f"\nSu parte lateral izquierda, cuenta con un color {lateral_izq['color']}:\n"
    descripcion += f" - Presenta una ventana {lateral_izq['ventana']['tipo']} hecha en vidrio templado y su marco en madera de roble.\n"
    descripcion += f" - En el vidrio está plasmado el logo {lateral_izq['logo_floristeria']['nombre']}, que cuenta con un diseño {lateral_izq['logo_floristeria']['diseño']}, acompañado de {lateral_izq['logo_floristeria']['elementos_visuales']}.\n"

    # Parte Posterior
    descripcion += f"\nSu parte posterior, cuenta con un color {posterior['color']}:\n"
    descripcion += f" - Incluye una ventana superior que funciona como {posterior['tragaluz']}.\n"
    descripcion += f" - Una matera colgante hecha de {posterior['matera_colgante']['material']}, contando con unas medidas de ({posterior['matera_colgante']['medidas_cm']}), que en su interior alberga {posterior['matera_colgante']['flores_cantidad']} girasoles.\n"

    # Lateral Derecho
    descripcion += f"\nSu parte lateral derecha, cuenta con un color {lateral_der['color']}:\n"
    descripcion += f"  - Tiene una ventana {lateral_der['ventana']['tipo']} que permite ver la decoración de las {lateral_der['ventana']['decoracion_interior']} en su interior.\n"
    descripcion += f"  - Ademas, cuenta con diferentes estantes de exhibición ({len(lateral_der['estantes']['tipos'])}):\n"
    descripcion += f"    - Espiral: Contiene {lateral_der['estantes']['detalle']['espiral']['items']} materos con {lateral_der['estantes']['detalle']['espiral']['tipo_flor']}.\n"
    descripcion += f"    - Alargado: De madera de roble, exhibe {lateral_der['estantes']['detalle']['alargado']['items']} orquídeas.\n"
    descripcion += f"    - Sencillo: Con 4 apartados que albergan un total de {lateral_der['estantes']['detalle']['sencillo']['cantidad_rosas']} rosas.\n"

    # Resumen de Flores
    descripcion += f"\n--- RESUMEN DE FLORES EN LA TIENDA: ---\n"
    descripcion += f" - Total de Flores en Exhibición (Interior): {total_flores_venta} unidades (Lirios, Orquídeas y Rosas).\n"
    descripcion += f" - Total de Flores de Decoración (Exterior): {total_flores_jardin} girasoles.\n"
    descripcion += "-------------------------------------------------------------------------------------------------------------------------------------"

    return descripcion

print(generar_descripcion_completa())
```
