```python
class Edificio:
    def __init__(self, material_paredes, diseno_seguridad, material_techo):
        self.material_paredes = material_paredes
        self.diseno_seguridad = diseno_seguridad
        self.material_techo = material_techo
        self.color_techo = "amarillo debido al desgaste por el sol y la lluvia"

    def describir_base(self):
        """Describe la estructura fundamental del edificio."""
        return (f"Estructura Base: 4 paredes Hechas de {self.material_paredes}, "
                f"con un diseo de seguridad de {self.diseno_seguridad}.\n"
                f"Techo: De tamao grande realizado con {self.material_techo}. "
                f"Cuentan con un color {self.color_techo}.")

class Puerta:
    def __init__(self, color, ancho_cm, alto_cm, material, herrajes):
        self.color = color
        self.medidas = f"({ancho_cm} x {alto_cm} cm)"
        self.material = material
        self.herrajes = herrajes

    def describir(self):
        return (f"La puerta de entrada es de color {self.color}, "
                f"contando con unas medidas de {self.medidas}, "
                f"fabricada en {self.material} con {self.herrajes}.")

class Ventana:
    def __init__(self, tipo, marco, descripcion_adicional):
        self.tipo = tipo
        self.marco = marco
        self.descripcion_adicional = descripcion_adicional

    def describir(self):
        return (f"Presenta una ventana {self.tipo} {self.descripcion_adicional} "
                f"y su marco en {self.marco}.")

class Matera:
    def __init__(self, material, ancho_cm, alto_cm, flor, cantidad):
        self.material = material
        self.medidas = f"({ancho_cm} cm de ancho x {alto_cm} cm de alto)"
        self.flor = flor
        self.cantidad = cantidad

    def describir(self):
        return (f"Una matera colgante hecha de {self.material}, "
                f"contando con unas medidas de {self.medidas}, "
                f"que en su interior alberga {self.cantidad} {self.flor}.")

class Estante:
    def __init__(self, tipo, material, descripcion_flores, cantidad_total):
        self.tipo = tipo
        self.material = material
        self.descripcion_flores = descripcion_flores
        self.cantidad_total = cantidad_total

    def describir(self):
        material_desc = f"De {self.material}, " if self.material else ""
        return (f"- {self.tipo}: {material_desc}{self.descripcion_flores}.")



class Floristeria(Edificio): #Herencia
    def __init__(self, nombre, material_paredes, diseno_seguridad, material_techo, color_fachada_principal, color_lateral_izq,
                 color_posterior, color_lateral_der, puerta, ventana_izq, ventana_post, matera_post, ventana_der, lista_estantes):

        super().__init__(material_paredes, diseno_seguridad, material_techo)

        self.nombre = nombre
        self.color_fachada_principal = color_fachada_principal
        self.color_lateral_izq = color_lateral_izq
        self.color_posterior = color_posterior
        self.color_lateral_der = color_lateral_der
        self.puerta = puerta
        self.ventana_izq = ventana_izq
        self.ventana_post = ventana_post
        self.matera_post = matera_post
        self.ventana_der = ventana_der
        self.estantes = lista_estantes

    def calcular_total_flores_interior(self):
        """Suma las flores de todos los estantes interiores."""
        total = 0
        for estante in self.estantes:
            total += estante.cantidad_total
        return total

    def pintar(self, parte, nuevo_color):
        """Pinta una parte especfica de la floristera."""
        if parte == "fachada principal":
            self.color_fachada_principal = nuevo_color
        elif parte == "laterales":
            self.color_lateral_izq = nuevo_color
            self.color_lateral_der = nuevo_color
        elif parte == "puerta":
            self.puerta.color = nuevo_color
        else:
            print(f"No se puede pintar '{parte}'.")
            return

        print(f"La/El {parte} ha sido pintada(o) de color {nuevo_color}.")

    def describir_arquitectura(self):
        """Genera la descripcin arquitectnica completa."""

        print(f"--- Descripcin Arquitectnica de la {self.nombre} ---\n")

        print(f"{super().describir_base()}\n")

        print(f"La fachada principal, cuenta con un color {self.color_fachada_principal}:")
        print(f"- {self.puerta.describir()}\n")

        print(f"Su parte lateral izquierda, cuenta con un color {self.color_lateral_izq}:")
        print(f"- {self.ventana_izq.describir()}\n"
              f"  - En el vidrio est plasmado el logo Flor by Ista, que cuenta con un diseo moderno y minimalista "
              "en letras cursivas, acompaado de un racimo de 3 flores y dos hojas.\n")

        print(f"Su parte posterior, cuenta con un color {self.color_posterior}:")
        print(f"- {self.ventana_post.describir()}")
        print(f"- {self.matera_post.describir()}\n")

        print(f"Su parte lateral derecha, cuenta con un color {self.color_lateral_der}:")
        print(f"- {self.ventana_der.describir()}")
        print(f"- Ademas, cuenta con diferentes estantes de exhibicin ({len(self.estantes)}):")
        for estante in self.estantes:
            print(f"    {estante.describir()}")

        total_interior = self.calcular_total_flores_interior()
        total_exterior = self.matera_post.cantidad

        print("\n--- RESUMEN DE FLORES EN LA TIENDA: ---")
        print(f"- Total de Flores en Exhibicin (Interior): {total_interior} unidades (Lirios, Orqudeas y Rosas).")
        print(f"- Total de Flores de Decoracin (Exterior): {total_exterior} girasoles.")
```
